### <문제 - 김하리>
정답인 핸드폰 번호는 다음과 같습니다
- real_num = “010-4657-8976”

1. 자신의 11자리 핸드폰 번호(010-0000-0000)를 입력하고,
2. 다음 조건에 따라 일정 값을 출력하세요.
    - 번호가 1개 이상, 10개 이하로 다를 경우에는 "X개 틀렸어요."를 (여기서 X개란 실제 틀린 개수를 의미합니다),
    - 모두 다를 경우에는 "전화 잘못 거셨어요."를,
    - 모두 일치한다면 "왜 전화하셨어요?"를 출력

````python
# split >> list 이용하기
real_num = "010-4657-8976"
real_num_list = real_num.split('-') #[010, 4657, 8976]

phone_num = input("Enter your phone number> ") 
phone_num_list = phone_num.split('-') #[010, 0000, 0000]

num = 0 #틀린 개수

for i in range(len(phone_num_list)): 
    for j in range(len(phone_num_list[i])):
        if phone_num_list[i][j] != real_num_list[i][j]:
            num += 1

if num == 0:
    print("왜 전화하셨어요?")
elif num == 11:
    print("전화 잘못 거셨어요.")
else:
    print(f"{num}개 틀렸어요.")
````

### <문제 - 임세은>
1. 다음과 같이 리스트를 설정한다. (리스트명은 변경 가능)
- 음료 = [‘사이다’, ‘콜라’, ‘제로콜라’, ‘식혜’, ‘커피]
- 가격 = [800, 1000, 1500, 2000, 2500]

2. 음료 명과 번호, 가격을 사용자에게 출력해 보여주고, 금액을 입력받은 후 음료의 가격이 입력받은 금액보다 적은 경우 구입 완료 메세지와 함께 잔액을 출력한다.
- 음료 메뉴는 for문을 사용해 출력한다.
- 사용자의 메뉴 입력 및 잔액 판단은 while문을 이용한다.

3. 사용자가 입력한 음료의 가격이 잔액보다 많은 경우, 잔액 부족 메세지를 출력한다.

4. 사용자가 종료를 위해 0을 입력하면 프로그램을 종료하고, 남은 잔액을 500원과 100원으로 나눠 반환한다. (ex: 잔액 800원 -> 500원 1개, 100원 3개)
- 프로그램 종료 후 남은 잔액 반환은 re_money() 함수를 지정해 사용한다. (re_money는 parameter, return이 있는 함수이다. return 시 500원 개수, 100원 개수를 모두 반환한다.)
- 500원은 정수 형태로 반환하되, 100원은 50원 잔액의 발생 여부를 고려해 float 형태로 반환한다.

````python
menu = ["사이다", "콜라", "제로콜라", "식혜", "커피"]
price = [800, 1000, 1500, 2000, 2500]

print("< 음료 자판기 프로그램 >")
print("==> 판매하는 음료 메뉴 <==")
for i in range(5):
    print(f"    {i+1} : {menu[i]} {price[i]}")
    
money = int(input("\n돈을 투입하세요 : "))

menu_num = None
while menu_num != 0: 
    menu_num = int(input("\n메뉴 선택(종료 : 0) :  ")) #*while 위가 아니라 안에 넣자!!!
    if money < price[menu_num-1]:
        print("잔액 부족")
        print(f"잔액: {money}")
    else:
        print(f"\n{menu[menu_num-1]} 구입 완료")
        money = money - price[menu_num-1]
        print(f"잔액: {money}") 
print(f"\n => 잔액 {money} 반환")

def re_money(money):
    num5 = money // 500
    num1 = (money % 500) / 100
    return [num5, num1]
              
print(f"\n500원 : {re_money(money)[0]} 개")
print(f"100원 : {re_money(money)[1]} 개")
print("음료 자판기 프로그램 종료.")
````

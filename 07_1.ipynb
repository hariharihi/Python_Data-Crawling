## CSS Selector 사용법
- select() 내 ```태그``` OR ```CSS class``` 이름을 넣어주면 된다
- 결과값: 리스트 형태로 반환

### 초간단 예제1: 태그 선택

````python
import requests 
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html') 
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('li') #tag 선택
for item in items:
    print(item.get_text())
````

````python
import requests 
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html') 
soup = BeautifulSoup(res.content, 'html.parser')
item = soup.select_one('ul#dev_course_list > li.course.paid') #가장 처음에 매칭되는 것 가져옴
print(item.get_text()) 
````

### 추출한 것에서 추출하기: find() VS select()
- 추출한 데이터에서 find() 사용해 재추출 가능
- find() 사용해 추출한 데이터를 select() 사용해 재추출 가능 >> 역도 성립!

````python
from bs4 import BeautifulSoup
import requests 

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html') 
soup = BeautifulSoup(res.content, 'html.parser')
item = soup.select('td') 
for item in items:
    print(item.get_text()) 
````

````python
from bs4 import BeautifulSoup
import requests 

res = requests.get('https://davelee-fun.github.io/blog/crawl_test_css.html') 
soup = BeautifulSoup(res.content, 'html.parser')
item = soup.select('tr td') #이게 더 안전
for item in items:
    print(item.get_text()) 
````

````python
# programming 익히기

from bs4 import BeautifulSoup
import requests

res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('tr') #3개의 행
for item in items:
    columns = item.select('td') #3개의 열 >> 리스트 형태
    row_str = ''
    for column in columns: 
        row_str += ', ' + column.get_text()
    print(row_str[2:])
````

````python
# programming 익히기

from bs4 import BeautifulSoup
import requests

res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.find_all('tr')
for item in items:
    columns = item.select('td') 
    row_str = ''
    for column in columns: 
        row_str += ', ' + column.get_text()
    print(row_str[2:])
````
